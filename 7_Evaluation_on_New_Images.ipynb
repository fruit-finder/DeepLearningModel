```python
import cv2
import numpy as np
import math
def get_preprocessed_imgs(dir_path):
  lt=[]
  imgs=[]
  label=[]
  imgs_lt=os.listdir(dir_path)
  imgs_lt=sorted(imgs_lt)
  print(imgs_lt)
  pass
  for i in range(len(imgs_lt)):
    image=cv2.imread(os.path.join(dir_path,imgs_lt[i]))
    try:
      resized_img=cv2.resize(image,(164,164))
      lt.append(resized_img)
      imgs.append(imgs_lt[i])
      label.append(int((i)/4))
    except:
      print(imgs_lt[i])
  return np.array(lt),imgs,np.array(label)
```

```python
image_data,image_names,target=get_preprocessed_imgs('/content/drive/MyDrive/All fruits/Evaluation fruit')
image_data.shape,len(image_names),target.shape
```
