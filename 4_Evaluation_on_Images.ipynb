```python
model=keras.models.load_model('/content/drive/MyDrive/All fruits/temp2/FruitsNewLatestBestModel2.h5')
```

```python
img_path='/content/drive/MyDrive/Resume /mango-fruit-1.jpg'
img=cv2.imread(img_path)
img.shape
```

```python
resized_img=cv2.resize(img,(164,164))
resized_img=np.expand_dims(resized_img,axis=0)
resized_img.shape
```

```python
y_pred=model.predict(resized_img)
y_pred
```

```python
#Finding top 3 predicted classes
num_of_classes=3
pred_classes=(-y_pred[0]).argsort()[:num_of_classes]
pred_classes
```

```python
for label in pred_classes:
  print(label_fruit_map[label],y_pred[0][label])
```
